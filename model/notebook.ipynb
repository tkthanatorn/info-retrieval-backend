```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/recipes.parquet")
df.head()
```

```python
import re

def camel_to_snake(name):
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()

df.columns = [camel_to_snake(col) for col in df.columns]
df.head()
```

```python
df["recipe_id"] = df["recipe_id"].astype(int)
df.head()
```

```python
df = df[["recipe_id", "name", "description", "recipe_instructions", "recipe_ingredient_quantities", "recipe_ingredient_parts", "images"]]
df.head()
```
